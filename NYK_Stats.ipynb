```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.basketball-reference.com/teams/NYK/2023.html"

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
table = soup.find('table', id = 'per_game')
```

```python
per_game_titles = table.find_all('th', class_ = 'poptip')
```

```python
table_titles = [title.text for title in per_game_titles]

print(table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = table_titles)

df = df.drop('Rk', axis=1)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    table_data = [data.text for data in row_data]
    
    length = len(df)
    df.loc[length] = table_data
    
```
