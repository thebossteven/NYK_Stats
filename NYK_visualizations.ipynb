```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"/Users/stephennunez/repos/NYK_Stats/NYK_Stats.csv")
plt.style.use('seaborn-v0_8-pastel')
```

```python
#group by wether the player is a starter or not(games started greater than 50)

df['Is Starter'] = ['Starter' if x > 50 else 'Not Starter' for x in df['GS']]
df.set_index('Player')

df_starter = df.groupby('Is Starter')
```

```python
df_starter.get_group('Starter')
```

```python
df_starter.plot(kind = 'bar', y = 'PTS', x = 'Player')
plt.tight_layout()
plt.savefig("Player_points.png")
```

```python
df_starter.plot(kind = 'barh', y = 'FG%', x = 'Player')
plt.tight_layout()
plt.savefig("Player_FieldGoal.png")
```

```python
knicks_ppg = df['PTS'].sum()
df_starter.sum().plot(kind = 'pie', y = 'PTS', autopct = '%1.0f%%')
plt.tight_layout()
plt.savefig("Points.png")
```
